## **Preprocesamiento**

Debido a la gran cantidad de columnas (575), se ha decidido combinar las variables que se repiten mensualmente, con el objetivo de reducir la compolejidad del análisis y modelamiento de los datos.

```python
IndRev= [col for col in monopoly.columns if col.startswith('IndRev')]
monopoly.drop(IndRev, axis='columns', inplace=True)
```

Se ha decidido eliminar IndRev, ya que es una variable que no presenta documentación y obstaculiza la reducción de columnas al no poder promediarse ni sumarse debido al tipo de dato (object) que almacena

```python
# flags mensuales
flg_cols = [col.replace('T01', '') for col in monopoly.columns if (col.endswith('01') and col.startswith("Flg"))] # cambiar por regex
print(flg_cols)
print(len(flg_cols))
```

Se obtiene el prefijo de todas las columnas de tipo Flg.

```python
# sumar columnas de flags mensuales
for prefix in flg_cols:
  cols_to_sum = [col for col in monopoly.columns if col.startswith(prefix)]
  monopoly[prefix + 'sum'] = monopoly[cols_to_sum].sum(axis=1) # nueva columna con suma de flags
  monopoly.drop(cols_to_sum, axis="columns", inplace=True) # eliminar columnas procesadas
```

Se suman las columnas de tipo Flg, que representan actividad en un mes X. Obteniendo por cada tipo de variable repetida mensualmente una sola variable anual que suma el total de meses en que se efectuo la actividad X. La desición de sumar estos valores se respalda en que son variables binarias, por lo que no tiene sentido utlizar un promedio.

```python
# lista de variables mensuales numéricas
month_cols = [col.replace('T01', '') for col in monopoly.columns if col.endswith('01')]
print(month_cols)
print(len(month_cols))
```

Prefijo de todas las variables mensuales que no son de tipo Flg.

```python
# calcula el promedio anual de las columnas mensuales
for col_name in month_cols:
  cols_to_averrage = [col for col in monopoly.columns if col.startswith(col_name)] # columnas a promediar
  monopoly[cols_to_averrage] = monopoly[cols_to_averrage] / 12 # división por un escalar
  monopoly[col_name + 'avg'] = monopoly[cols_to_averrage].sum(axis=1)
  monopoly.drop(cols_to_averrage, axis="columns", inplace=True) # se eliminan las columnas procesadas
```

Se hace un promedio anual por cada variables que se repite mensualmente, exluyendo las variables de tipo Flg, tratadas en los pasos anteriores.

**Tratamiento de datos nulos**

```python
df_length = monopoly.shape[0] # cantidad de registros
na_sum = monopoly.isna().sum() # serie con total de nulos
have_na = na_sum > 0 # condicion: nulos > 0
na_sum[have_na] # serie filtrada
```

```python
df_length = monopoly.shape[0] # cantidad de registros
null_pct = monopoly.isna().sum() / df_length * 100 # total de nulos en porcentaje
have_nulls = null_pct > 0 # condicion pct de nulos mayor a 0
null_pct[have_nulls] # serie filtrada
```

```python
monopoly = monopoly.drop(columns='Unnamed: 574') # eliminación variable Unnamed
```

Se elimina la variable Unnamed debido a que solo presenta datos nulos

```python
monopoly['CambioPin'] = monopoly['CambioPin'].fillna(0) # nulos se reemplazan por 0
```

CambioPin presenta datos nulos en todos los registros que no son 1, por lo que se reemplazaron los datos nulos por 0, indicando que esos registros no presentan cambio de pin.

```python
monopoly['Region'] = monopoly['Region'].fillna(monopoly['Region'].mode().iloc[0]) # se imputan los nulos por la moda
```

Para los nulos de región, estos se imputaron por la moda ya que son pocos registros de una variable categórica con una moda muy predominante.

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5)
monopoly = pd.DataFrame(imputer.fit_transform(monopoly), columns = monopoly.columns)
```

Para evitar la perdida del 26% de los registros, los valores faltantes se imputaron utilizando KNN, que mostró mejores resultados que otras formas de imputación.

**Tratamiento valores negativos**

```python
filter = monopoly.min() < 0
negative_vars = monopoly.min()[filter].index.to_list()
monopoly[negative_vars] = monopoly[negative_vars].abs()
```

Se tratan los valores negativos utilizando el valor absoluto, ya que estos se encontraban en variables relacionadas a deudas, por lo que puede que los valores negativos se deban a un error de signo al ingresar los datos o a unir datos de diferentes fuentes.

**Outliers**

Debido al a la forma en que se distribuyen los datos en la mayoría de variables se ha decidido no eliminar ni imputar outliers, ya que la mayor parte de los datos de las diferentes variables con outliers, tienen una muy alta concentración de sus datos en 0., por lo que los pocos datos que varían se ven reflejados como outliers sin necesariamente ser datos erroneos.

Matriz de correlación del dataset.

```python
#Correlación de Pearson | CUENTAS CLI
corr_matrix = monopoly.corr()
corr_matrix['Cuentas'].sort_values(ascending=False)
```

```python
#Correlación de Pearson | Cuenta Corriente
corr_matrix = monopoly.corr()
corr_matrix['Ctacte'].sort_values(ascending=False)
```

```python
# Correlación de Pearson
corr_matrix = monopoly.corr()

# Filtrar correlaciones mayores a 0.7 (o cualquier otro umbral deseado)
correlations_above_07 = corr_matrix[corr_matrix > 0.7]

# Eliminar la diagonal principal (correlaciones perfectas)
np.fill_diagonal(correlations_above_07.values, np.nan)

# Obtener las correlaciones mayores a 0.7
high_correlations = correlations_above_07[correlations_above_07 > 0.7]

# Crear un gráfico de calor con el estilo de gradiente de color 'coolwarm'
plt.figure(figsize=(10, 8))
sbn.heatmap(high_correlations.dropna(how='all', axis=1).dropna(how='all'),
            annot=True, cmap='coolwarm', fmt='.2f',
            linewidths=0.5, linecolor='black', cbar=False)

plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(rotation=0, fontsize=10)
plt.title('Correlaciones mayores a 0.7', fontsize=12)
plt.show()
```
